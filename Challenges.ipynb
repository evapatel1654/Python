```python
#Format a string value
raw_username = " Py_Thon-ROCKS123!!! "

username = raw_username.strip()
username = username.lower()
username = username.replace("-", "_")
username = username.replace("!", "")
has_digit = any(char.isdigit() for char in username)
contains_python = "python" in username
first_8_chars = username[:8]

print(f"Cleaned Username: {username}")
print(f"Contains a digit: {has_digit}")
print(f"Contains 'python': {contains_python}")
print(f"First 8 characters: {first_8_chars}")
```

```python
#Work with Lambda
students = [
    ('alice', 34),
    ('bob', 22),
    ('diana', 19)
]
sorted_students = sorted(students, key=lambda x: x[1], reverse=True)

print(sorted_students)
```

```python
#Manipulate Tuples
students = [
    ("Alice", 20, "A"),
    ("Bob", 19, "B"),
    ("Charlie", 21, "A"),
    ("Diana", 20, "C")
]

students_with_A = [s for s in students if s[2] == "A"]
print("Students with A:", students_with_A)

names = [s[0] for s in students]
print("Names only:", names)

updated_students = [
    (name, age, "B") if age == 20 else (name, age, grade)
    for name, age, grade in students
]

print("Updated students:", updated_students)
```

```python
#Manipulate Sets
attendees = [
    "Alice", "Bob", "Alice", "Ethan",
    "Bob", "Diana", "Ethan", "Frank"
]

attendees_set = set(attendees)
print("Total attendees:", len(attendees_set))

if "Alice" in attendees_set:
    print("Alice is going!")

attendees_set.add("Grace")
attendees_set.discard("Bob")
attendees_set.discard("Bob")

print("Final attendees set:", attendees_set)
```
